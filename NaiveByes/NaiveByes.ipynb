```python
import pandas as pd
```

```python
data=pd.read_csv('Iris.csv')
data.head(3)
```

```python
data.isnull().sum()
```

```python
data['Species'].unique()
```

```python
print(data.columns)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X=data.drop(columns=['Id','Species'],axis=1)
```

```python
Y=data['Species']
```

```python
X=scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
```

```python
model.fit(X_train,Y_train)
```

```python
Predict_X=model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
As=accuracy_score(Predict_X,Y_train)
print(As)
```

```python
PredictTest=model.predict(X_test)
```

```python
score=accuracy_score(PredictTest,Y_test)
print(score)
```
